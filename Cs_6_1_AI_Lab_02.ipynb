Lab Task

1. Append() function


```python
Fruits = ["Apple",150,"Banana",200,"Orange",250,"Mango",300]

Fruits.append("Graps")

Fruits.append(400)

print(Fruits)
```

2. Clear() function


```python
Fruits = ["Apple",150,"Banana",200,"Orange",250,"Mango",300]

Fruits.clear()

print(Fruits)
```

3. Copy() function

```python
Fruits = ["Mango",150,"Banana",200,"Apple",300,"Oranges",400]

Vegetables = []

Vegetables = Fruits.copy()

print(Vegetables)
```

4. Count() function

```python
Fruits = ["Mango",150,"Banana",200,"Banana",300,"Oranges",400]
total = Fruits.count("Banana")
print("Total: ",total)
```

5. extend() function

```python
Fruits = ["Mango",150,"Banana",200,"Apple",300,"Oranges",400]
Vegetables = ["Tomato",400]
Fruits.extend(Vegetables)

print(Fruits)
```

6. Index() function


```python
Fruits = ["Mango",150,"Banana",200,"Apple",300,"Oranges",400]

print(Fruits[4])
```

7. Insert() function

```python
Fruits = ["Mango",150,"Banana",200,"Apple",300,"Oranges",400]

Fruits.insert(2,"Graps")

print(Fruits)
```

8. Pop() function

```python
Fruits = ["Mango",150,"Banana",200,"Apple",300,"Oranges",400]

Fruits.pop(3)

print(Fruits)
```

9. Remove() function

```python
Fruits = ["Mango",150,"Banana",200,"Apple",300,"Oranges",400]

Fruits.remove("Apple")

print(Fruits)
```

10. Reverse() function

```python
Fruits = ["Mango",150,"Banana",200,"Apple",300,"Oranges",400]

Fruits.reverse()

print(Fruits)
```

11. Sort() function

```python
Fruits = ["Banana","Apple","Orange","Mango"]

Fruits.sort()

print(Fruits)
```

Bonus Point- GPA Calculator in python

```python
print("Welcome to my GPA calculator")

total_courses = int(input("Enter your total number of courses: "))

total_points = 0
total_credits = 0

for i in range(total_courses):
    print("\nCourse", i + 1)

    course = input("Enter your course name: ")
    credit_hours = int(input("Enter your credit hours: "))
    obtain_marks = int(input("Enter your obtained marks: "))

    if obtain_marks >= 80:
        grade_points = 4.0
    elif obtain_marks >= 75:
        grade_points = 3.7
    elif obtain_marks >= 70:
        grade_points = 3.3
    elif obtain_marks >= 65:
        grade_points = 3.0
    elif obtain_marks >= 60:
        grade_points = 2.7
    elif obtain_marks >= 55:
        grade_points = 2.3
    elif obtain_marks >= 50:
        grade_points = 2.0
    elif obtain_marks >= 45:
        grade_points = 1.7
    else:
        grade_points = 0.0

    total_points += grade_points * credit_hours
    total_credits += credit_hours

gpa = total_points / total_credits
print("\nYour GPA is:", round(gpa, 2))
```
